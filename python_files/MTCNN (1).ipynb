```python
from mtcnn import MTCNN
import cv2
import glob
import numpy as np
import pandas as pd
```

```python
detector = MTCNN()
```

```python
noface_err_imgs = list()
gif_err_imgs = list()
```

```python
def save_cropimg(path):
    try:
        # mtcnn 수행
        detector = MTCNN()    

        img = cv2.cvtColor(cv2.imread(path), 
                           cv2.COLOR_BGR2RGB)
        face_info = detector.detect_faces(img)
        if len(face_info) == 0 :
            noface_err_imgs.append(path)
        else:
            face_bb_coor = face_info[0]['box']

            x, y, w, h = face_bb_coor


            #  Bounding Box 좌표대로 Image Crop % save
            img = cv2.imread(path)
            crop_img = img[y:y+h, x:x+w]
            cv2.imwrite('./crop_imgs/' + path.split('/')[-1], crop_img)
    except:
        gif_err_imgs.append(path)
```

```python
all_img_list = glob.glob('./imgs/*')
all_img_list
```

```python
len(all_img_list)
```

```python
for path in all_img_list:
    save_cropimg(path)
```

```python
noface_err_imgs
```

```python
gif_err_imgs
```

```python
err_imgs = noface_err_imgs + gif_err_imgs

for path in err_imgs:
    save_cropimg(path)
```

```python
import shutil
shutil.make_archive('crop_imgs', 'zip', './crop_imgs')
```
