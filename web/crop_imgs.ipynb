```python
import math
from sklearn import neighbors
import os
import os.path
import sys
import pickle
from PIL import Image, ImageDraw, ImageFont
import face_recognition
from face_recognition.face_recognition_cli import image_files_in_folder
```

```python
err = []
cutfailed = []

for i in range(len(test_list)):
    try:
        # img_path = 'C:/Users/user/Desktop/summerdev/summerdev/public/celeb\\(G)I-DLE 민니1.jpg'
        img_path = test_list[i]

        img = Image.open(img_path)
        img.convert('RGB')

        img_show = Image.open(img_path).convert("RGB")
        iw, ih = img_show.size

        img = face_recognition.load_image_file(img_path)

        face_locations = face_recognition.face_locations(img)
        bt, br, bb, bl = face_locations[0]
        bw, bh = (br - bl, bb - bt)
        bcx, bcy = (bl + bw // 2, bt + bh // 2)
        # print('원본 바운딩박스 t, r, b, l ', bt, br, bb, bl)
        # print('원본 바운딩박스 크기 w, h ', bw, bh)
        
        # l_length = bcx
        # r_length = iw - bcx
        # t_length = bcy
        # b_length = ih - bcy
        

        nl, nt, nr, nb = (bl + bw // 8, bt, br - bw // 8, bb)
        nw, nh = (nr - nl, nb - nt)
        ncx, ncy = (nl + nw // 2, nt + nh // 2)


        l_length = ncx * 4 // 3
        r_length = (iw - ncx) * 4 // 3
        t_length = ncy
        b_length = ih - ncy

        lens = [l_length, r_length, t_length, b_length]
        min_len = min(lens)

        if min_len == b_length:
            crop_box = (ncx - (b_length) * 3 // 4, ncy - b_length, ncx + (b_length) * 3 // 4, ncy + b_length)

        elif min_len == t_length:
            crop_box = (ncx - t_length * 3 // 4, 0, ncx + t_length * 3 // 4, 2 * ncy)    

        elif min_len == l_length:
            l_length = ncx
            crop_box = (0, ncy - l_length * 4 // 3, 2 * ncx, ncy + l_length * 4 // 3)
            
        elif min_len == r_length:
            r_length = iw - ncx
            crop_box = (2 * ncx, ncy - r_length * 4 // 3, iw, ncy + r_length * 4 // 3)    

        cbl, cbt, cbr, cbb = crop_box
        # print('crop_box', crop_box)
        # print('이미지 자를 crop_box w, h ', cbr-cbl, cbb-cbt)

        # # Crop the image
        cropped_img = img_show.crop(crop_box)
        cropped_img.save('C:/Users/user/Desktop/summerdev/summerdev/public/crop_celeb/' + img_path.split('\\')[-1])

        if cbl < 0 or cbt < 0 or cbr > iw or cbb > ih:
            cutfailed.append(img_path)
    except:
        err.append(test_list[i])
        pass
print('err ', err)
print('cutfailed ', cutfailed)
```
